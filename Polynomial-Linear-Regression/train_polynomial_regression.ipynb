# Polynomial Regression

## Importing the libraries

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing the dataset

```
df = pd.read_csv('Position_Salaries.csv')
x = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
reg_pol = PolynomialFeatures(degree=4)
x_pol = reg_pol.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_pol, y)
```

## Visualising the Linear Regression results

```
plt.scatter(x, y, color='blue')
plt.plot(x, lin_reg.predict(x), color='red')
plt.title('Linear Regression')
plt.xlabel('level')
plt.ylabel('salary')
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(x, y, color='blue')
plt.plot(x, lin_reg_2.predict(x_pol), color='red')
plt.title('Polynomial Regression')
plt.xlabel('level')
plt.ylabel('salary')
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

## Predicting a new result with Linear Regression

```
lin_reg.predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```
lin_reg_2.predict(reg_pol.fit_transform([[6.5]]))
```

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Data sample
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2.1, 4.5, 9.8, 16.2, 25.3])

# Fit model untuk derajat yang berbeda dan plot
plt.scatter(x, y, color='blue')

for degree in range(1, 6):  # Derajat polinomial dari 1 hingga 5
    poly = PolynomialFeatures(degree)
    x_poly = poly.fit_transform(x)
    model = LinearRegression()
    model.fit(x_poly, y)
    y_poly_pred = model.predict(x_poly)
    plt.plot(x, y_poly_pred, label=f'Degree {degree}')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Fitting Polynomial Model')
plt.legend()
plt.show()
```
